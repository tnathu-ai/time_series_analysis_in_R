```python
#Tasks - Module 3 -- Time Series Analysis

#Loading necessary packages

import pandas as pd
import numpy as np
from statsmodels.tsa.arima_process import ArmaProcess
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import acf, pacf
import seaborn as sns
```

```python
#Task 1

#a) Generating simulation of AR1 with phi=0.6
plt.subplot(2,1,1)
ar1 = np.array([1, -0.6]) #The phi value is entered with the opposite sign 
ma1 = np.array([1]) #ma is not assigned any theta values since only AR is present
AR_object1 = ArmaProcess(ar1, ma1)
simulated_data_1 = AR_object1.generate_sample(nsample=100)
plt.plot(simulated_data_1)

#Creating the ACF and PACF plots
sm.graphics.tsa.plot_acf(simulated_data_1, lags=20);
sm.graphics.tsa.plot_pacf(simulated_data_1, lags=20);

# Set of possible models : {AR(1), ARMA(1,2)}
```

```python
#b) Generating simulation of AR1 with phi=-0.6
plt.subplot(2,1,1)
ar2 = np.array([1, 0.6]) #The phi value is entered with the opposite sign 
ma2 = np.array([1]) #ma is not assigned any theta values since only AR is present
AR_object2 = ArmaProcess(ar2, ma2)
simulated_data_2 = AR_object2.generate_sample(nsample=100)
plt.plot(simulated_data_2)

#Creating the ACF and PACF plots

sm.graphics.tsa.plot_acf(simulated_data_2, lags=20);
sm.graphics.tsa.plot_pacf(simulated_data_2, lags=20);

# Set of possible models : {ARMA(1,1), AR(1) = ARMA(1,0)} 
```

```python
#c) Generating simulation of AR2 with phi=(0.6,-0.6)
plt.subplot(2,1,1)
arparams = np.array([0.6,-0.6]) #save the two phi coefficients in a variable
ar3 = np.r_[2, -arparams] #The phi values are entered with the opposite sign 
ma3 = np.array([1]) #ma is not assigned any theta values since only AR is present
AR_object3 = ArmaProcess(ar3, ma3)
simulated_data_3 = AR_object3.generate_sample(nsample=100)
plt.plot(simulated_data_3)

#Creating the ACF and PACF plots
sm.graphics.tsa.plot_acf(simulated_data_3, lags=40);
sm.graphics.tsa.plot_pacf(simulated_data_3, lags=40);

# Set of possible models : {AR(2)} 
```
